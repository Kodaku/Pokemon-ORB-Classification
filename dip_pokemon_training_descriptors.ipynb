```python
import glob
import cv2
from matplotlib import pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
import os
import pandas as pd
from skimage.transform import resize
from skimage.io import imread
%matplotlib inline
from google.colab import drive
```

```python
drive.mount("/content/drive", force_remount=True)
train_path = "/content/drive/Shareddrives/PokémonAdami/PureDataset/PokémonTrain"
descriptors_path = "/content/drive/Shareddrives/PokémonAdami/PureDataset/ResizedTrainDescriptors"
```

```python
pokemon_names = os.listdir(train_path)
print(pokemon_names), len(pokemon_names)
```

```python
#si ferma qui
import csv 
for pokemon_name in pokemon_names:
  new_path = train_path + "/" + pokemon_name
  print(f"Image Path: {new_path}")
  desc_path = descriptors_path + "/" + pokemon_name
  print(f"Descriptor Path: {desc_path}")
  if not os.path.isdir(desc_path):
    os.mkdir(desc_path)
  train_imgs_names = os.listdir(new_path)
  descriptors = []
  for img_name in train_imgs_names:
    train_image = cv2.imread(new_path + "/" + img_name)
    train_image = cv2.resize(train_image, (256, 256))
    orb = cv2.ORB_create()
    kp, desc = orb.detectAndCompute(train_image, None)
    descriptors.append(desc)
  counter = 0
  for descriptor in descriptors:
    try:
      array = np.array([range(descriptor.shape[1])])
      with open(descriptors_path + "/" + pokemon_name + "/" + pokemon_name + "_" + str(counter) + ".csv","w+") as my_csv:
          newarray = csv.writer(my_csv,delimiter=',')
          newarray.writerows(array)
          newarray.writerows(descriptor)
          print(f"Written {pokemon_name} decriptors number {counter}")
          counter += 1
    except:
      continue
```
