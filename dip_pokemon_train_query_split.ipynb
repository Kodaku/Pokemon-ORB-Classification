```python
import glob
import cv2
from matplotlib import pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
import os
import pandas as pd
from skimage.transform import resize
from skimage.io import imread
%matplotlib inline
from google.colab import drive
```

```python
drive.mount("/content/drive", force_remount=True)
data_path = "/content/drive/Shareddrives/PokémonAdami/PureDataset/Images"
train_path = "/content/drive/Shareddrives/PokémonAdami/PureDataset/PokémonTrain"
query_path = "/content/drive/Shareddrives/PokémonAdami/PureDataset/PokémonQueries"
```

```python
pokemon_names = os.listdir(data_path)
```

```python
print(pokemon_names)
```

```python
#Looping through the folders
for pokemon_name in pokemon_names:
  pokemon = pokemon_name.split(".")[1]
  print("Writing " + pokemon + " images")
  new_path = data_path + "/" + pokemon_name
  pokemon_images_names = os.listdir(new_path)
  pokemon_images = np.zeros((len(pokemon_images_names), 80, 80, 3), dtype=np.uint8)
  i = 0
  #reading images for a single pokémon
  for image_name in pokemon_images_names:
    image_path = new_path + "/" + image_name
    try:
      image = cv2.imread(image_path)
      image = cv2.resize(image, (80, 80))
      pokemon_images[i] = image
    except:
      pass
    i += 1
  #shuffling the images and split them in train and test images
  np.random.seed(42)
  np.random.shuffle(pokemon_images)
  total_length = pokemon_images.shape[0]
  test_size = 11
  train_size = total_length - test_size
  train_images = pokemon_images[:train_size]
  test_images = pokemon_images[train_size:]
  #ensuring no image is left behind
  assert len(train_images) + len(test_images) == len(pokemon_images)
  #saving training images
  i = 0
  for train_image in train_images:
    new_path = query_path + "/" + pokemon
    # pokemon_name = pokemon.split('.')[1]
    if not os.path.exists(new_path):
      os.mkdir(new_path)
    cv2.imwrite(new_path + "/" + pokemon + str(i) + "_query.jpg", train_image)
    i += 1
  #saving test images
  i = 0
  for test_image in test_images:
    new_path = train_path + "/" + pokemon
    # pokemon_name = pokemon.split('.')[1]
    if not os.path.exists(new_path):
      os.mkdir(new_path)
    cv2.imwrite(new_path + "/" + pokemon + str(i) + "_train.jpg", test_image)
    i += 1
```
